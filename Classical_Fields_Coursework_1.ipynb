```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
```

```python
#a
x = np.linspace(-2.5, 4, 100)
y=0
h = (np.exp(-x**2 - y**2)) + 0.5*np.exp(-(x-1.5)**2 - 2*y**2)
plt.plot(x, h)
```

```python
#b
w = np.arange(-2, 3, 0.05)
[x, y]= np.meshgrid(w, w)
h = (np.exp(-x**2 - y**2)) + 0.5*np.exp(-(x-1.5)**2 - 2*y**2)

fig = plt.figure(figsize=(9,7))
ax = fig.gca()

l = np.arange(0.1, 1.1, 0.1)
contourplot = ax.contour(x, y, h, levels=l)
ax.set_aspect('equal')
```

```python
#d
w = np.arange(-2, 3, 0.05)
[x, y]= np.meshgrid(w, w)
h = (np.exp(-x**2 - y**2)) + 0.5*np.exp(-(x-1.5)**2 - 2*y**2)

fig = plt.figure(figsize=(9,7))
ax = fig.gca()

l = np.arange(0.1, 1.1, 0.1)
contourplot = ax.contour(x, y, h, levels=l)
ax.set_aspect('equal')

w1 = np.arange(-2, 3, 0.5)
[x, y]= np.meshgrid(w1, w1)

u = (-2*x*np.exp(-x**2 -y**2)) -(x-1.5)*np.exp(-(x-1.5)**2 - 2*y**2)
v = (-2*y*np.exp(-x**2 -y**2)) -2*y*np.exp(-(x-1.5)**2 - 2*y**2)

plt.quiver(x, y, u, v, width=0.003, angles='xy', scale_units='xy', scale = 3)
ax.margins(0.1)
ax.set_aspect('equal')
```

```python
#e
w = np.arange(-2, 3, 0.05)
[x, y]= np.meshgrid(w, w)
h = (np.exp(-x**2 - y**2)) + 0.5*np.exp(-(x-1.5)**2 - 2*y**2)

fig = plt.figure(figsize=(9,7))
ax = fig.gca()

l = np.arange(0.1, 1.1, 0.1)
contourplot = ax.contour(x, y, h, levels=l)
ax.set_aspect('equal')

v = np.arange(-2, 3, 0.5)
[x, y] = np.meshgrid(v, v) # Set up grid

h = (np.exp(-x**2 - y**2)) + 0.5*np.exp(-(x-1.5)**2 - 2*y**2)

[gradphiv, gradphiu] = np.gradient(h, 0.5)

ax.quiver(x, y, gradphiu, gradphiv, width=0.003, angles='xy',
           scale_units='xy', scale=3)
ax.margins(0.1)
ax.set_aspect('equal')
```

```python
#%% f
w = np.arange(-2, 3, 0.05)
[x, y]= np.meshgrid(w, w)
h = (np.exp(-x**2 - y**2)) + 0.5*np.exp(-(x-1.5)**2 - 2*y**2)

fig = plt.figure(figsize=(9,7))
ax = fig.gca()

l = np.arange(0.1, 1.1, 0.1)
contourplot = ax.contour(x, y, h, levels=l, cmap=cm.rainbow)
ax.set_aspect('equal')

for level in l:
    contour_points = contourplot.collections[l.tolist().index(level)].get_paths()
    for contour_path in contour_points:
        vertices = contour_path.vertices
        x_contour, y_contour = vertices[:, 0], vertices[:, 1]
        skip = 10 # Skip every 10th point to reduce arrows
        u = (-2*x_contour[::skip]*np.exp(-x_contour[::skip]**2 - y_contour[::skip]**2)) - (x_contour[::skip]-1.5)*np.exp(-(x_contour[::skip]-1.5)**2 - 2*y_contour[::skip]**2)
        v = (-2*y_contour[::skip]*np.exp(-x_contour[::skip]**2 - y_contour[::skip]**2)) - 2*y_contour[::skip]*np.exp(-(x_contour[::skip]-1.5)**2 - 2*y_contour[::skip]**2)
        ax.quiver(x_contour[::skip], y_contour[::skip], u, v, width=0.003, angles='xy', scale_units='xy', scale=3, color='black')

ax.margins(0.1)
ax.set_aspect('equal')
```
